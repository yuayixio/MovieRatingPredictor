```python
from time import sleep
from imdb import IMDb

ia = IMDb()
```

```python
from tqdm import tqdm
```

```python
movies = pd.read_csv('../../data/preprocessed/movies_id_updated.csv')
movies.head()
```

```python
keywords_dict = {}
for movie_index in tqdm(movies['imdbID']):
    sleep(1)
    try:
        keywords_dict[movie_index] = ia.get_movie_keywords(movie_index[2:])['data']['keywords']
    except:
        keywords_dict[movie_index] = ''   
```

```python
#write dictionary in csv
import csv
with open('../../data/tuning/keywordDict.csv', 'w') as csv_file:  
    writer = csv.writer(csv_file)
    for row in keywords_dict.items():
        writer.writerow(row)
```
